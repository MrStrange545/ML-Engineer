## 2.1.Representation Learning
**Goal:** Given a set of `data points`, our goal is to `understand` something `useful` about these data points.

Data Points:
- Vectors in $\mathbb{R}^d$

- E.g. $\begin{bmatrix} Height \\ weight \\ age \end{bmatrix} \in \mathbb{R}^3$

To explain what is the meaning of `understand` and `useful` we will fix one theme, that we will call running theme.

Running theme: `COMPREHENSION IS COMPRESSION`.
Comprehend -> Learn\Understand

Problem: 
- Input: $\{x_1, x_2, x_3, ..., x_n \} x_i \in \mathbb{R}^d $
    - d is the number of features
- Output: some `Compressed` representation of the dataset.

**What is the meaning of compressed representation of the dataset?**
Let see this with the help of an example.

Example:

Data points
$\left\{ \begin{bmatrix} -7 \\ -14 \end{bmatrix} , \begin{bmatrix} 2.5 \\ 5 \end{bmatrix} , \begin{bmatrix} 0.5 \\ 1 \end{bmatrix}, \begin{bmatrix} 0 \\ 0 \end{bmatrix} \right\}$

**How many real numbers are needed to store this dataset?**
- Naive answer is 8.
- Here we may use representative point

Representative: 
$\begin{bmatrix} 1 \\ 2 \end{bmatrix} \in \mathbb{R}^2$

Coefficient: 
$\{-7,2.5,0.5,0 \}$

So, the answer to previous question, that how many real numbers needed to store the above dataset is 6.

**How much do we save?**
- From this example, it may seem there are no great savings as we only go from 8 to 6 points.
- But if we have  billion data points then we have to store in 2 billion real numbers but with our approach we can just store it in (1 billion + 2) numbers.

**NOTE:**
- Using representative and coefficients we can `RECONSTRUCT` the dataset exactly.



````python
X = np.array([[-7,-14],
              [2.5,5],
              [0.5,1],
              [0,0]])

x = np.linspace(-10,10,100)
y = 2*x

plt.plot(x,y,c="b",  label="y=2x")
plt.scatter(X[:,0], X[:,1], c = "r",marker = "X",s=100)

plt.xlim(-7.5,5)
plt.ylim(-15,10)
plt.legend()
plt.show()
````

**Can we only take $\begin{bmatrix} 1 \\ 2 \end{bmatrix} \in \mathbb{R}^2$ as representative?**
- No, we can take more than one representative.

Another Rep.
$\begin{bmatrix} \frac{1}{\sqrt{5}} \\ \frac{2}{\sqrt{5}} \end{bmatrix} \in \mathbb{R}^2$

Then coefficient will change accordingly
$\{-7\sqrt{5}, 2.5\sqrt{5}, 0.5\sqrt{5}, 0 \}$

**NOTE:**
- Any vector along the the Blue line can be chosen as representative. Except $\begin{bmatrix} 0 \\ 0 \end{bmatrix}$

Summary:
```
Original Representation: # real numbers = 2n
Compressed Representation: # real numbers = n + 2

In General,

Original Representation: # real numbers = d * n
Compressed Representation: # real numbers = n + d
```
We were able to able to compress the data points because they were having some relation. 

**What if they don' thave any relation or In geometric sense what if data points doesn't lie on this line?**

````python
X = np.array([[-7,-14],
              [2.5,5],
              [0.5,1],
              [0,0],
              [2,-4]])

x = np.linspace(-10,10,100)
y = 2*x

plt.plot(x,y,c="b",  label="y=2x")
plt.scatter(X[:,0], X[:,1], c = "r",marker = "X",s=100, label=["X1","X2","X3","X4","X5"])

for i, (xi, yi) in enumerate(X):
    plt.annotate(f'X{i+1}', (xi,yi), textcoords="offset points", xytext=(0,10), ha="center")
plt.xlim(-7.5,5)
plt.ylim(-15,10)
plt.legend()
plt.show()
````

In this scenario, we have to make a `proxy data point` for the data point which is not on the line.

Proxy data point: A data point which is on the line but it is the which comes in place of another data point which do not lie on the line.

we set proxy data point in such a way that we lose the minimum value.

**Who can "pretend" to be  a proxy for X5 along the blue line?** <br>
Ans: Projection of X5 onto blue line.

**How to find the projection of X5 on the line?**
Let' assume that $ w = \begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$ is the representative of this line. and we have the point let's say $x = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$

Now, we can see $\begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$ vector as the basis for this line and hence $ c \cdot \begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$ will act as a proxy point.

So, we need to find the value of `c` that loss is minimum. 

We can frame this as optimizatin problem.
$$\text{minimise c }: length^2(\text{ error vector } )$$
$$\text{minimise c }: length^2(\begin{bmatrix} x_1-cw_1 \\ x_2-cw_2 \end{bmatrix})$$

$$\text{minimise c }: (x_1-cw_1)^2 + (x_2-cw_2)^2 $$

you can differentiate the above equation wrt to c and set it to zero to find the minimum value of c.
$$c^* = \frac{x_1w_1+x_2w_2}{w_1^2+w_2^2} \text{      (scalar)}$$

$c^*$ is the proxy point.

How do we interpret the above result of c?
$$c = \frac{\vec{x}\cdot \vec{w}}{||w||^2} = \frac{x^Tw}{||w||^2}$$

So the projection or the vector on the blue is 
$$= c \cdot \begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$$
$$=  \frac{x^Tw}{||w||^2} \cdot \begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$$

**NOTE:**
To make the computations easier for us, We can always pick w such that length of $\begin{bmatrix} w_1 \\ w_2 \end{bmatrix} = 1$. 

So, we are going to use representatives which have length = 1.

**what is the advantage of above approach?** <br>
(proxy point) $c^* =(x^Tw)\begin{bmatrix} w_1 \\ w_2 \end{bmatrix}$

Given a line, we can find the proxy of the data point on the line, but In practice we are not given a line, we are just given a dataset, therefore, we also need to find the line.

**Goal:** Develop a way to find a `compressed representation` of data when data points not-necessarily fall on line.

In real world, no dataset will perfectly or even partially fit to some line (linear model). So, for that, we are always going to find proxies of those points.

Now, we can plot any line and find proxies of the original data points on that line.

All line will give me same compression ratio that is 50% in our case, but then also One line will be the best fit line, why? 

Because the reconstruction error is the least in that specific line.

**Goal:** GIven a dataset, find the line that has least reconstruction error.

Dataset: $\{x_1, x_2, x_3, ..., x_n \} x_i \in \mathbb{R}^d $

ERROR(line, dataset)
$$=\sum^n_{i=1} error(line, x_i)$$
$$=\sum^n_{i=1} length^2(x_i-(x_i^Tw)w)$$
$$=\sum^n_{i=1} ||x_i-(x_i^Tw)w||^2$$

our goal is to minimize the w(line) and we will take average of the error.

Let's make it a function of w.
$$f(w)= \frac{1}{n} \sum^n_{i=1} ||x_i-(x_i^Tw)w||^2$$

Remember, $x_i-(x_i^Tw)w$ is a vector in $\mathbb{R}^d$ because x in $\mathbb{R}^d$ and w is also in $\mathbb{R}^d$.

The line mentioned above is represented using `w` such that $||W||=1$

**NOTE:** $||Z||^2 = Z^TZ$

Let's do the calculation.
$$f(w)= \frac{1}{n} \sum^n_{i=1} ||x_i-(x_i^Tw)w||^2$$

$$f(w)= \frac{1}{n} \sum^n_{i=1} (x_i-(x_i^Tw)w)^T \cdot(x_i-(x_i^Tw)w) $$

$$f(w)= \frac{1}{n} \sum^n_{i=1} [ x_i^Tx_i - (x_i^Tw)^2 - (x_i^Tw)^2 + (x_i^Tw)^2 1  ] $$

$$f(w)= \frac{1}{n} \sum^n_{i=1} [ x_i^Tx_i - (x_i^Tw)^2 ]$$

Now, we want to minimise f(w) but we have a constant in our function, so we can neglect that constant and focus on w part.

$$g(w)= \frac{1}{n} \sum^n_{i=1} -(x_i^Tw)^2$$

Our goal is to minimise g(w) over w, but we have negative sign in the g(w) so what we will do is not minimise the w but maximise the g(w) but after removing the sign.

$$\text{[max over w ]} g(w)= \frac{1}{n} \sum^n_{i=1} (x_i^Tw)^2$$

$$= \frac{1}{n} \sum^n_{i=1} (w^Tx_i)(x_i^Tw)$$

$$= \frac{1}{n} \sum^n_{i=1} w^T(x_i,x_i^T)w$$

$w \text{ and } w^T$ does not depend on i so we can pull them out.

$$= w^T (\frac{1}{n} \sum^n_{i=1} x_i,x_i^T)w$$

This structure $ (\frac{1}{n} \sum^n_{i=1} x_i,x_i^T)$ is a $d \times d$ matrix. It is also known as `covariance matrix`.

Let consider this covariance matrix as `C`. then,

$$ \text{max w   } g(w) = w^TCw $$ only when $||w||^2 = 1$.

We want to find direction w which maximises this by linear form .

**Solution to this problem:** w is the eigenvector corresponding to the maximum eigenvalue of C.

---

Now, we will see, the thing which we are calling the error "might not the the error".

$$x \in \mathbb{R}^d \rightarrow$$ 
Find w 
$$(x^Tw)\cdot w \rightarrow$$
Residue\error 
$$x_i-(x_i^Tw)\cdot w$$

and we are saying this $x_i-(x_i^Tw)\cdot w$ might not be the error `but has information`.

**POSSIBLE ALGORITHM** <br>
Input: $\{x_1, x_2, x_3, ..., x_n \} x_i \in \mathbb{R}^d $

Here we have to make the origin as the centre of our datapoints. <br>
So, we have to find the centered dataset.

First, find the mean: $\mu = \frac{1}{n} \sum_{i=1}^nx_i$ <bR>
then, $x_i = x_i - \mu \space \forall \space i$


-> Find "best" line $w_1 \in \mathbb{R}^d $ <bR>
-> Replace $x_i \leftarrow x_i-(x_i^Tw)\cdot w$ <br>
-> Repeat to obtain $w_2$ and so on.

**What is the need of centering the dataset?**

````python
X = np.array([[-2,-14],
              [7.5,5],
              [5.5,1],
              [5,0],
              [4,-6]])

x = np.linspace(-10,10,100)
y = 2*x

plt.plot(x,y,c="b",  label="y=2x")
plt.scatter(X[:,0], X[:,1], c = "r",marker = "X",s=100, label=["X1","X2","X3","X4","X5"])

for i, (xi, yi) in enumerate(X):
    plt.annotate(f'X{i+1}', (xi,yi), textcoords="offset points", xytext=(0,10), ha="center")

plt.legend()
plt.show()
````

As, you can see our "blue line" is not the best line because data points are in the same direction as the blue line is but then we have to move the origin to the mean of these data points.

### 2.1.1 Principal component analysis
Dataset:  $\{x_1, x_2, x_3, ..., x_n \} x_i \in \mathbb{R}^d $
$$w_1 = \underset{w, ||w||^2 =1}{argmax} \space w^TCw $$
>argmax: it means that w which maximises the objective function.

$$C = \frac{1}{n}\sum^n_{i=1}(x_ix_i^T)$$

after first iteration, 

$$x_1^\prime \rightarrow x_1-(x_1^Tw_1)w_1$$

$$x_2^\prime \rightarrow x_2-(x_2^Tw_1)w_1$$
$$x_3^\prime \rightarrow x_3-(x_3^Tw_1)w_1$$

$$...$$

$$x_n^\prime \rightarrow x_n-(x_n^Tw_1)w_1$$

New Dataset: $\{x_1^\prime, x_2^\prime, x_3^\prime, ..., x_n^\prime \} x_i \in \mathbb{R}^d $

Now, what is $w_2$
$$w_2 = \underset{w, ||w||^2 =1}{argmax } \space w^TC^\prime w $$
$C^\prime$ is the co-variance matrix of the new dataset.
$$C^\prime = \frac{1}{n}\sum^n_{i=1}(x_i^\prime (x_i^\prime)^T)$$

Now, we have two lines, one from the original dataset, that means the line which best fits the dataset in terms of least reconstruction error. and another line which fits the errors in terms of their least reconstruction error.

**Question:** Is there any relation btw $w_1$ and $w_2$? <br>
Observations: 
- All residues are orthogonal to $w_1$.
- Any line which minimises sum of errors wrt residues must also be orthogonal to $w_1$.
$$w_2^Tw_1 = 0$$

Here, we were doing in two dimensions, so only two w but what if we have more than 2 dimensions let's say 100 dimensions.

Then we will not stop just after 2 lines, then we have to find residue of residue and so on till the final residue becomes 0.

So,, By continuing this procedure we get, 
$$\{w_1, w_2, ..., w_d \} \text{ s.t. } ||w_k||^2 =1 \space \forall \space k \text{ and } w_i^Tw_j = 0 \space \forall i \ne j$$

Basically, we get $w_3$ which is orthogonal to $w_1, w_2, w_4, ..., w_d$ (every other vector except $w_3$.

In other terms, we get a set of `Orthonomal vectors`( a set which contains vectors whose norm is 1 and perperndicular to each other).

---
**Residues after round 1:**
$$\{x_1-(x_1^Tw_1)w_1, ...., x_n-(x_n^Tw_1)w_1 \text{ all vectors are in } \mathbb{R}^d\}$$

- $w_1^Tw_2 = 0$
- $w_2 \rightarrow$ Best line that fits residues.

**Residues after round 2:**
$$\{(x_1-(x_1^Tw_1)w_1) - ((x_1-(x_1^Tw_1)w_1^Tw_2)w_2, .... \}$$

Let's simplify the above vector
$$x_1-(x_1^Tw_1)w_1 - (x_1^Tw_2)w_2, ...$$ 
because all other terms will become zero as $w_1Tw_2 = 0$

**Residues after d-rounds:**
$$\forall i \space x_i-\left( (x_i^Tw_1)w_1 + (x_i^Tw_2)w_2 + ... + (x_i^Tw_d)w_d  \right) = 0 $$

$$\forall i \space x_i = \left( (x_i^Tw_1)w_1 + (x_i^Tw_2)w_2 + ... + (x_i^Tw_d)w_d  \right)$$

all, we have done is "change of basis".

**What have we gained?** <br>
- If our data lives in a "low" dimensional linear sub-space, then residues becomes 0 much earlier than d rounds.

Example: say dataset is such that after three rounds the residues becomes 0. Dataset initially was in 100 dimension $\{x_1, x_2, x_3, ..., x_n \} x_i \in \mathbb{R}^100 $.

What does it mean to say "residues becomes zero after three rounds" it means -> we have found $w_1, w_2, w_3 $ such that $\forall \space i \space x_i = (x_i^Tw_1)w_1 + (x_i^Tw_2)w_2 + (x_i^Tw_3)w_3$ well $\{ w_1, w_2, w_3  \} \in \mathbb{R}^100$ are the lines which we have found through our algorithm.

Let's see the representatives and coefficients: <br>
Rep: $\{ w_1, w_2, w_3  \} \in \mathbb{R}^100$ common for the entire dataset<br>
Co-efficients per data point: $x_i \rightarrow \left[ x_i^Tw_1,x_i^Tw_2, x_i^Tw_3\right] \in \mathbb{R}^3$ datapoint specific

well,
```
Orginally: 100 * n real numbers
Now: 100 * 3 + 3n

In general,
orginal: d*n
after algo: d*k + k*n , k is the number of rounds after which residue becomes zero.
```
**Question:** What if data "approximately" lie in a low-dimensional space? <br>

Now, we are defining the threshold. Here, threshold mean how much error we are accepting to bear or tolerate.

By pythagorus theorem, (for $\mathbb{R}^2$ )<br>
square of the length of the data point = square of the error + square of the projection.

$$\text{ for any w } \in \mathbb{R}^d, \text{ s.t. } ||w||^2 =1$$
$$ \forall \space i \space ||x_i||^2 = \underset{projection}{||(x_i^Tw)w||^2} + \underset{error}{||x_i-(x_i^Tw)w||^2}$$

average, 

$$\frac{1}{n}\sum^n_{i=1}||x_i||^2 = \frac{1}{n}\sum^n_{i=1}||x_i-(x_i^Tw)w||^2 + \frac{1}{n}\sum^n_{i=1}||(x_i^Tw)w||^2$$

we want this quantity $\frac{1}{n}\sum^n_{i=1}||(x_i^Tw)w||^2$ as large as possible as that error can be minimum.

"larger the value of $\frac{1}{n}\sum^n_{i=1}(x_i^Tw^2)$, the better the fit."

![K rounds needed only](attachments/03.png)

after K rounds, we will consider the information to be the error before that we believe it will retrieve or contain about 95% or whatever the threshold we will decide.

---
**ENTER LINEAR ALGEBRA**
$$\underset{w ,||w||^2 = 1}{max} w^TCw \text{ where } C = \frac{1}{n}\sum^n_{i=1}x_ix_i^T$$

C is a covariance matrix

Soln: $w_1$ which is the eigenvector corresponding to the largest eigenvalue of C. It comes from `Hilbert min-max theorem`.

In fact, $\{w_1 ,...., w_d \}$, the eigenvectors of c form an orthonormal basis.

$w_k \rightarrow$ best line one can obtain in round k.

**What do eigenvalues of C mean?** <br>
we know, 
$$Cw_1 = \lambda_1w_1$$
$$w_1^TCw_1 = w_1^T(\lambda_1w_1) = \lambda_1 \text{ because } w_1^Tw_1 = ||w|| = 1$$
$$\lambda_1 = w_1^TCw_1 = w_1^T\left(\frac{1}{n}\sum^n_{i=1}x_ix_i^T\right)w_1$$
$$\lambda_1 = \frac{1}{n}\sum^n_{i=1}\left(x_i^Tw_1 \right)^2$$


where, $Cw_1$ is the eigen vector and $\lambda_1$ is the largest eigenvalue corresponding to the vector $w_1$.

**Rule of thumb**: for # dimensions <br>
$\left( \frac{\sum^L_{i=1}\lambda_i(C)}{\sum^d_{i=1}\lambda_i(C)}\right) >= 0.95 \text{ ususally in practice }$

we have seen the coefficient are:
$$\{(x_i^Tw), ... ,(x_i^Tw)\}$$

Average: 
$$\frac{1}{n} \sum_{i=1}^n (x_i^T)w$$
we are assuming that dataset is centered, so for a centered dataset, mean = 0 
$$\frac{1}{n} \sum_{i=1}^n (x_i^T)w = 0$$

For the same set of values, let's look at the variance, how much the spread of values is: <br>
$$\frac{1}{n} \sum_{i=1}^n(x_i^Tw-mean)^2$$
but we know, that mean is zero so,  variance is 
$$\frac{1}{n} \sum_{i=1}^n(x_i^Tw)^2$$

Now, we come to know that the term which we are trying to maximise earlier to minimise the error is the "variance" of the coefficients.

`ERROR MINIMIZATION ON CENTERED DATASET` $\equiv$ `VARIANCE MAXIMIZATION`

Geometrically, it means that if the projections of data-points on the line are closer to each other, it will be very difficult to distinguish at reconstruction time, whereas if there is high spread among the projections of data points then that line will give least reconstruction error. <br>
You can even see this phenomena in the following picture.
![why variance maximization](attachments/04.png)

when data points are projected over the blue line they are closer to each other, but when data points are projected on green line they are spreaded so green line is a better fit than blue line.

We want directions where projections don't "crowd up" i.e. variance is not small.

---

Now, we will summarize this algorithm with the help of an example:

we have the dataset which contains height and weight of the persons.
we are assuming `Height gives some information about weight` i.e. height increases so the weight increases.

If I choose to represent height and weight as two numbers then one number gives some information about the second number.

![height and weight example](attachments/05.png)

let denote this line with $w_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$

But we have to normalize it, so $w_1 = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}$

so this w is the direction, so what we will store is dot product of this direction and data point. 
$$x_i^Tw_1$$
which will be equal to (in our example)
$$\frac{height + weight} {\sqrt{2}}$$

Now, if I want to find the another line onto which I will project the error vectors then this line must be orthonormal to line $w_1$ and that is:
$$w_2 = \begin{bmatrix} \frac{-1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}$$

![height and weight example](attachments/06.png)

The two directions which we have found are not related to each other, one direction is not showing any trend or specific pattern change wrt change to other which means the directions $\frac{h+w}{\sqrt{2}}$ and $\frac{w-h}{\sqrt{2}}$ are `de-correlated`.

Every Direction gives you some new information that the previous directions are not giving you, you are trying to find those directions, So that is what our algorithm is essentially doing.

This algorithm is called `Principal component analysis`.

and, the $\{w_1, w_2, ..., w_k \}$ are called `Principal components`. here k is the number of rounds after which there is only noise or irrelevant information(less variance).

The first principal component $w_1$ captures the most variance, the second principal component $w_2$ captures the second most variance, and so on.

This is also the `DIMENSIONALITY REDUCTION` Technique.

PCA finds combination of features that are de-correlated [Independent of each other (loosely) ].

**<u>Standard PCA</u>**

<u>Steps involved in PCA </u><bR>
Step 1: Center the dataset <br>
Step 2: Calculate the covariance matrix of centered data <br>
step 3: compute the eigenvectors and eigenvalues of this covariance  matrix. <br>
step 4: sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues. <br>
step 5: Transform the origianl data by multiplying it with the selected eigenvectors(Principal components) to obtain a lower dimensinal representation. <br>

- n: no. of datapoints

Example: <br>

````python
# Code
# First, import the required package
import numpy as np
import matplotlib.pyplot as plt


# Have the dataset and print the dataset
X = np.array([(4,1),(5,4),(6,3),(7,4),(2,-1),(-1,-2),(0,-3),(-1,-4)]).T
print("\nActual Dataset:\n",X)


# calculate the mean of the dataset and print it
x_mean = X.mean(axis = 1)
print(f"\nmean of the dataset: {x_mean}\n")


# plot the dataset along with its mean
print("\nPlotting the original dataset\n")
plt.scatter(X[0, :],X[1, :])
plt.axhline(0, color="k")
plt.axvline(0, color="k")
plt.scatter(x_mean[0],x_mean[1], color="r")
plt.grid()
plt.show()


# Now Center the dataset, reassigning it, and printing it.
def center(X):
    return X- X.mean(axis = 1).reshape(2,1)
X_centered = center(X)
print(f"\nDataset after being centered:\n",X_centered,sep="\n")


# calculate the mean of centered dataset and print it
x_mean = X_centered.mean(axis = 1)
print(f"\nMean after centering the dataset: {x_mean}\n")


# Plotting the dataset along with its mean (after centering)
print("\nPlotting the centered dataset.\n")
plt.scatter(X_centered[0, :],X_centered[1, :])
plt.axhline(0, color="k")
plt.axvline(0, color="k")
plt.scatter(x_mean[0],x_mean[1], color="r")
plt.grid()
plt.show()


# Computing the covariance matrix
def covariance(X):
    return X @ X.T / X.shape[1]    
C = covariance(X_centered)
print(f"\nCovariance Matrix:\n",C)


# Find the principal components
def compute_pc(C):
    d = C.shape[0]
    eigval, eigvec = np.linalg.eigh(C)
    w_1, w_2 = eigvec[:, -1], eigvec[:, -2]
    return w_1, w_2
    
w_1, w_2 = compute_pc(C)

w_1 = w_1[:,np.newaxis]
w_2 = w_2[:,np.newaxis]

print(f"\nThe principal components are:",w_1,w_2,sep="\n")


# Reconstruction using the two PCs
n = X_centered.shape[1]
reconstructed_dataset = w_1 @ (X_centered.T @ w_1).reshape(1,n) + w_2 @ (X_centered.T @ w_2).reshape(1,n)
print(f"\nReconstructed dataset is:",reconstructed_dataset, sep="\n")


# The reconstruction error by first 
````
